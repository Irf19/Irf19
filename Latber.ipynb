```r
set.seed(1)
df <- data.frame(x = rnorm(20, 72, 8), 
                 y = rnorm(20, 40, 20),
                 Umpan_Fred = sample(c("On Target", "Mbuh Nandi"),
                               40, replace = TRUE))

ggplot(df) +
 annotate_pitch(colour = "white",
                fill = "chartreuse4") +
 geom_point(aes(x = x, y = y, fill = Umpan_Fred),
            shape = 21,
            size = 6) +
 coord_cartesian(xlim = c(15, 100))+ 
 theme_pitch() +
 theme(panel.background = element_rect(fill = "chartreuse4"))
```

```r
set.seed(1)
df <- data.frame(x = rnorm(4, 7, 2), 
                 y = rnorm(4, 2, 4),
                 Umpan_Fred = sample(c("On Target", "Mbuh Nandi"),
                               20, replace = TRUE))

ggplot(df) +
 annotate_pitch(colour = "white",
                fill = "chartreuse4") +
 geom_point(aes(x = x, y = y, fill = Umpan_Fred),
            shape = 21,
            size = 6) +
 coord_cartesian(xlim = c(15, 100))+ 
 theme_pitch() +
 theme(panel.background = element_rect(fill = "chartreuse4"))
```

```r
pass <- data.frame(x1 = c(5, 3, 6, 7),
                   x2 = c(5, 5, 6, 8),
                   y1 = c(5, 7, 4, 3),
                   y2 = c(7, 4, 3, 8))

ggplot(pass) +
 annotate_pitch(colour = "white",
                fill = "#3ab54a") +
  geom_segment(aes(x = x1, y = y1, xend = x2, yend = y2),
               arrow = arrow(length = unit(0.3, "cm"),
                             type = "closed")) +
 theme_pitch() +
 theme(panel.background = element_rect(fill = "#3ab54a"))
```
